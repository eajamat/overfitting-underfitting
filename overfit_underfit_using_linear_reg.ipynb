```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
#Generating random data (original data)
np.random.seed(1)
X = np.random.uniform(0, 1, (20,1))
y = X**2 + np.random.uniform(0, 0.1, (20, 1))
#y = X**2
X = np.sort(X, axis=0)
y = np.sort(y, axis=0)
```

```python
#plotting the original data
plt.plot(X, y, "ro", label="original data")
plt.show()
```

```python
#Fitting linear model
model_linear = LinearRegression()
model_linear.fit(X, y)
model_linear.score(X, y)
```

```python
#Fitting polynomial 2 model

X_poly_right = np.array([X, X**2]).T.reshape(-1, 2)
model_right = LinearRegression() 
model_right.fit(X_poly_right, y)
model_right.score(X_poly_right, y)
```

```python
#Fitting polynomial 8 model

X_poly_over = np.array([X, X**2, X**3, X**4, X**5, X**6, X**7, X**8]).T.reshape(-1, 8)
model_over = LinearRegression()
model_over.fit(X_poly_over, y)
model_over.score(X_poly_over, y)
```

```python
#Calculating accuracy

y_predict_linear = model_linear.predict(X)
y_predict_right = model_right.predict(X_poly_right)
y_predict_over = model_over.predict(X_poly_over)

y_linear_error = mean_squared_error(y_predict_linear, y)
y_right_error = mean_squared_error(y_predict_right, y)
y_over_error = mean_squared_error(y_predict_over, y)

print("Linear regression error is {}".format(y_linear_error))
print("Right fit regression error is {}".format(y_right_error))
print("Over fit regression error is {}".format(y_over_error))
```

```python
#Generating test data to plot the graphs
plot_predict = np.arange(0, 1, 0.01).reshape(-1,1)
plot_predict_right = np.array([plot_predict, plot_predict**2]).T.reshape(-1, 2)
plot_predict_over = np.array([plot_predict, plot_predict**2, plot_predict**3, plot_predict**4, plot_predict**5, 
                        plot_predict**6, plot_predict**7, plot_predict**8]).T.reshape(-1, 8)

y_plot_linear = model_linear.predict(plot_predict)
y_plot_right = model_right.predict(plot_predict_right)
y_plot_over = model_over.predict(plot_predict_over)
```

```python
#Plotting the undrfit graph

plt.plot(X, y, "ro", plot_predict, y_plot_linear)
plt.show()
```

```python
#Plotting the right fit graphs

plt.plot(X, y, "ro", plot_predict, y_plot_right)
plt.plot()
```

```python
#Plotting the overfitting graphs

plt.plot(X, y, "ro", plot_predict, y_plot_over)
plt.ylim(-0.3, 1.25)
plt.plot()
```

```python
X = np.random.uniform(0, 1, (100,1))
new_y = X ** 2
test_linear_error1 = mean_squared_error(y_plot_linear, new_y)
test_right_error1 = mean_squared_error(y_plot_right, new_y)
test_over_error1 = mean_squared_error(y_plot_over, new_y)

print("Linear regression error on test data is {}".format(test_linear_error1))
print("Right fit regression error on test data is {}".format(test_right_error1))
print("Over fit regression error on test data is {}".format(test_over_error1))
```
